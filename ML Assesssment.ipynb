Imagine you have a dataset where you have different Instagram features like u sername , Caption , Hashtag , Followers , Time_Since_posted , and likes , now your task is to predict the number of likes and Time Since posted and the rest of the features are your input features. Now you have to build a model which can predict the number of likes and Time Since posted. 



```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler


data = pd.read_csv('instagram_reach.csv')
```

```python
data
```

```python
data.head()
```

```python

X = data[['Followers', 'Time since posted']]  
y_likes = data['Likes']
y_time_since_posted = data['Time since posted']

```

```python

X_train, X_test, y_likes_train, y_likes_test, y_time_train, y_time_test = train_test_split(
    X, y_likes, y_time_since_posted, test_size=0.2, random_state=20
)

```

```python

from sklearn.preprocessing import MinMaxScaler

# Standardize the features using MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Create and train the model for predicting likes
model_likes = LinearRegression()
model_likes.fit(X_train_scaled, y_likes_train)

```

```python
# Create and train the model for predicting time since posted
model_time = LinearRegression()
model_time.fit(X_train_scaled, y_time_train)

```

```python

likes_predictions = model_likes.predict(X_test_scaled)
time_predictions = model_time.predict(X_test_scaled)

```

```python

likes_mse = mean_squared_error(y_likes_test, likes_predictions)
time_mse = mean_squared_error(y_time_test, time_predictions)

```

```python
likes_r2 = r2_score(y_likes_test, likes_predictions)
time_r2 = r2_score(y_time_test, time_predictions)

```

```python
print(f'Mean Squared Error for Likes: {likes_mse}')
print(f'R-squared for Likes: {likes_r2}')

print(f'Mean Squared Error for Time Since Posted: {time_mse}')
print(f'R-squared for Time Since Posted: {time_r2}')
```
